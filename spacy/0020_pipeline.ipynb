```python
%reset -f
import spacy
nlp = spacy.load("en_core_web_sm")
```

## Tokenization

Break words into individual words (or other characters) known as tokens.

```python
text = "This is a very simple sentence to be used while working with spaCy in JupyterLab and push changes to GitHub in 2026."
doc = nlp(text)
print("len(text):",len(text))
print("len(doc):",len(doc))
print("type(doc)",type(doc))
print([token.text for token in doc])
```

## Lemmatization

Turn words into their root form.

```python
lemmas = [token.lemma_ for token in doc]
print(lemmas)
```

## POS: Part of Speech

Assign grammatical categories—such as nouns, verbs, adjectives, or prepositions—to each individual word (or token) in a text based on its context.

```python
poses = [token.pos_ for token in doc]
print(poses)
```

## NER: Named Entity Recognition

Identify and classify named entities: people, organizations, locations, ...

```python
print(["{0}|{1}".format(token.text,token.label_) for token in doc.ents])
```

## Displacy - display

Enhanced visualization of the output.

```python
from spacy import displacy
displacy.render(doc,style='ent',jupyter=True)
```

## DP: Dependency Parser

Examine the dependencies between the words of a sentence to analyze its grammatical structure.

Ref: https://universaldependencies.org/u/dep/

```python
displacy.render(doc,style='dep',jupyter=True)
```
