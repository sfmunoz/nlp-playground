# NLTK simple example

Ref: https://www.nltk.org/

Downloaded data default PATH seems to be `${HOME}/nltk_data`

```python
%reset -f
import nltk
nltk.download('punkt_tab')  # to ${HOME}/nltk_data; ref: https://www.nltk.org/data.html
nltk.download('averaged_perceptron_tagger_eng')
sentence = "Hello world... how are you feeling today?"
tokens = nltk.word_tokenize(sentence)
tagged = nltk.pos_tag(tokens)
tagged[0:6]
```
