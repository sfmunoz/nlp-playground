## Tokenization

```python
import nltk
nltk.download('punkt_tab')  # to ${HOME}/nltk_data; ref: https://www.nltk.org/data.html
text = "This is a very simple sentence!"
tokens = nltk.word_tokenize(text)
print(tokens)
```
