# Pandas Plot Simple

Simple [Pandas](https://pandas.pydata.org/) plot example

```python
%reset -f
%matplotlib inline

import pandas as pd

s = pd.Series({1:18,2:19,3:11,4:14,5:16,6:14,7:15})
print(s.describe())
ax = s.plot()
ax.set_title("pd.Series()")
ax.legend(labels=["simple series"])
ax.grid(color='#c0c0c0',linestyle='--',linewidth=1)
```
