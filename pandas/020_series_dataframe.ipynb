## Series (simple)

```python
steps = pd.Series(range(4,40,7))
print("type(steps) ...",type(steps))
print(steps)
```

## DataFrame

```python
data = {
    "name": ["lemon","orange","lettuce"],
    "grams": [90, 140, 450],
    "color": ["yellow","orange","green"],
}
df = pd.DataFrame(data)
print("type(df) ...........",type(df))
print("type(df['name']) ...",type(df['name']))
print(df)
```
