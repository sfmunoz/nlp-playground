# Universal functions (ufunc)

- https://numpy.org/doc/stable/reference/ufuncs.html
- https://numpy.org/doc/stable/reference/generated/numpy.ufunc.html#numpy.ufunc
- https://numpy.org/doc/stable/reference/ufuncs.html#available-ufuncs

```python
%reset -f
import numpy as np
arr = np.arange(0,20)
# np.<ufunc>.accumulate(array)
arr2 = np.add.accumulate(arr)
# np.<ufunc>.reduce(array)
arr3 = np.add.reduce(arr)
print("arr:",arr)
print("np.add.accumulate(arr):",arr2)
print("np.add.reduce(arr):",arr3)
```
