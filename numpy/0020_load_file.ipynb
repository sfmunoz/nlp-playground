# Load file example

```python
%reset -f
import numpy as np
filename = "0020_load_file/data1.txt"
data = np.loadtxt(filename)
print("==== data ====")
print(data)
ids, feat_a, feat_b, feat_c = np.loadtxt(filename, unpack=True)
print("==== ids ====")
print(ids)
print("==== feat_b ====")
print(feat_b)
```

```python
c = ["zero","one","two","three","four","five","six","seven","eight","nine"]
def converter(x):
    try:
        return int(x)
    except:
        pass
    return c.index(x.lower())
filename = "0020_load_file/data2.txt"
data = np.loadtxt(filename,dtype=int,delimiter=",",converters=converter)
print("==== data ====")
print(data)
ids, feat_a, feat_b = np.loadtxt(filename,dtype=int,delimiter=",",converters=converter,unpack=True)
print("==== ids ====")
print(ids)
print("==== feat_a ====")
print(feat_a)
print("==== feat_b ====")
print(feat_b)
```
