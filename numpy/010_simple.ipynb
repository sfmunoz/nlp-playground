# NumPy simple example

```python
%reset -f
import numpy as np
arr = np.array([1,2,3,4,5])
print("arr =",arr)
arr += 3
print("arr += 3",arr)
arr *= 2
print("arr *= 2",arr)
print("arr[2:-1]",arr[2:-1])
arr += arr
print("arr += arr",arr)
```

```python
s = 8
arr = np.array([[j for j in range(i,i+s)] for i in range(0,s**2,s)])
print(arr)
print(arr[1:-1,1:-1])
```
