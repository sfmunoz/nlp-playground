# NumPy simple example

To get help run the following in a code cell:
- `help(...)`: e.g. `help(np)`, `help(np.array)`, ... from docstring
- `np?`, `np.array?`, ... from docstring
- `np??`, `np.array??`, ... to access the source code

```python
%reset -f
import numpy as np
```

```python
arr = np.array([1,2,3,4,5])
print("arr =",arr)
arr += 3
print("arr += 3",arr)
arr *= 2
print("arr *= 2",arr)
print("arr[2:-1]",arr[2:-1])
arr += arr
print("arr += arr",arr)
```

```python
s = 8
arr = np.array([[j for j in range(i,i+s)] for i in range(0,s**2,s)])
arr2 = arr[2:-2,1:-1]
print("arr.size:",arr.size)
print("arr.shape:",arr.shape)
print("arr.dtype:",arr.dtype)
print(arr)
print("arr2.size:",arr2.size)
print("arr2.shape:",arr2.shape)
print("arr2.dtype:",arr2.dtype)
print(arr2)
```

```python
arr = np.linspace(0,9.5,num=20)
print(arr)
arr2 = arr.reshape(2,2,5)
print(arr2)
```

```python
a1 = np.array([[1,1],[2,2]])
a2 = np.array([[3,3],[4,4]])
print("a1:",a1)
print("a2:",a2)
print("np.vstack()")
print(np.vstack((a1,a2)))
print("np.hstack()")
np.hstack((a1,a2))
```
